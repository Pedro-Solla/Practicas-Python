```python
import numpy as np
import pysindy as ps
```

```python
#Condiciones iniciales x_0=3 ; y_0=0.5
t = np.linspace(0, 1, 100)
x = 3 * np.exp(-2 * t)
y = 0.5 * np.exp(t)
X = np.stack((x, y), axis=-1)  # First column is x, second is y
```

```python
model = ps.SINDy()
model.fit(X, t=t, feature_names=["x", "y"])
model.print()
```
